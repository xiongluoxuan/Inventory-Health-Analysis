```python
# 先创建几个基础业务指标
df["Revenue"] = df["Units_Sold"] * df["Unit_Price"]
df["Inventory_Gap"] = df["Inventory_Level"] - df["Reorder_Point"]
df["Forecast_Gap"] = df["Inventory_Level"] - df["Demand_Forecast"]
df["Gross_Margin_Per_Unit"] = df["Unit_Price"] - df["Unit_Cost"]

# 查看新字段
display(df[[
    "SKU_ID",
    "Date",
    "Units_Sold",
    "Inventory_Level",
    "Reorder_Point",
    "Demand_Forecast",
    "Revenue",
    "Inventory_Gap",
    "Forecast_Gap",
    "Gross_Margin_Per_Unit"
]].head())
```

```python
sku_summary = df.groupby("SKU_ID").agg(
    Total_Units_Sold=("Units_Sold", "sum"),
    Total_Revenue=("Revenue", "sum"),
    Avg_Inventory=("Inventory_Level", "mean"),
    Avg_Reorder_Point=("Reorder_Point", "mean"),
    Avg_Inventory_Gap=("Inventory_Gap", "mean"),
    Avg_Forecast_Gap=("Forecast_Gap", "mean"),
    Avg_Lead_Time=("Supplier_Lead_Time_Days", "mean"),
    Avg_Unit_Cost=("Unit_Cost", "mean"),
    Avg_Unit_Price=("Unit_Price", "mean"),
    Avg_Gross_Margin_Per_Unit=("Gross_Margin_Per_Unit", "mean"),
    Promotion_Days=("Promotion_Flag", "sum"),
    Stockout_Days=("Stockout_Flag", "sum")
).reset_index()

display(sku_summary.head())
```

```python
def classify_sku(row):
    if row["Avg_Forecast_Gap"] < 0:
        return "Risk"
    elif row["Avg_Inventory_Gap"] < 50:
        return "Watch"
    elif row["Avg_Lead_Time"] > 10:
        return "Watch"
    else:
        return "Healthy"

sku_summary["Health_Status"] = sku_summary.apply(classify_sku, axis=1)

display(sku_summary[[
    "SKU_ID",
    "Total_Units_Sold",
    "Total_Revenue",
    "Avg_Inventory_Gap",
    "Avg_Forecast_Gap",
    "Avg_Lead_Time",
    "Health_Status"
]].head())
```

```python
# 1. 统计不同健康状态的 SKU 数量
print("SKU 健康状态数量：")
print(sku_summary["Health_Status"].value_counts())

# 2. 查看 Watch / Risk SKU
attention_skus = sku_summary[
    sku_summary["Health_Status"].isin(["Watch", "Risk"])
].copy()

# 3. 按收入从高到低排序，优先关注业务影响大的 SKU
attention_skus = attention_skus.sort_values(
    by="Total_Revenue",
    ascending=False
)

display(attention_skus[[
    "SKU_ID",
    "Health_Status",
    "Total_Units_Sold",
    "Total_Revenue",
    "Avg_Inventory_Gap",
    "Avg_Forecast_Gap",
    "Avg_Lead_Time",
    "Stockout_Days"
]].head(10))
```

```python
total_skus = sku_summary["SKU_ID"].nunique()
health_counts = sku_summary["Health_Status"].value_counts()

watch_count = health_counts.get("Watch", 0)
risk_count = health_counts.get("Risk", 0)
healthy_count = health_counts.get("Healthy", 0)

watch_ratio = watch_count / total_skus
risk_ratio = risk_count / total_skus

top_watch_skus = attention_skus.head(5)

print("SKU 总数:", total_skus)
print("Healthy 数量:", healthy_count)
print("Watch 数量:", watch_count)
print("Risk 数量:", risk_count)
print("Watch 占比:", round(watch_ratio, 2))
print("Risk 占比:", round(risk_ratio, 2))

display(top_watch_skus[[
    "SKU_ID",
    "Total_Revenue",
    "Avg_Lead_Time",
    "Avg_Inventory_Gap",
    "Avg_Forecast_Gap"
]])
```
